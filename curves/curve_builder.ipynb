```python
import pandas as pd 
import numpy as np
```

```python
class YieldCurve:
    def __init__(self , maturities , yields):
        self.maturities = np.array(maturities)
        self.yields = np.array(yields)

    def get_curve(self):
        return pd.Dataframe({'Maturities': self.maturities,
                             'Yields': self.yields
                             })
```
