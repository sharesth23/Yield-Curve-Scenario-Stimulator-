```python
import numpy as np 
from scipy.interpolate import interp1d
```

```python
class SwapCurve:
    def __init__(self ,tenors , rates):
        self.tenors = np.array(tenors)
        self.rates = np.array(rates)
        self.curve = interp1d(tenors, rates, kind="linear", fill_value="extrapolate")
```
